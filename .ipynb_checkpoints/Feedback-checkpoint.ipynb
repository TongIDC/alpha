```python
import pandas as pd
pd.set_option('display.max_rows',110)
```

```python
def calculate_error(Forecast,Real):
    error = ((Forecast-Real)/Real)*100
    return error
```

```python
check = pd.read_csv("check.csv")
```

```python
check["Gen_error"]=calculate_error(check["ForecastGen"],check["Generation"])
check["Load_error"]=calculate_error(check["ForecastCon"],check["Consumption"])
```

```python
check["Gen_error"].fillna(0 ,inplace=True)
```

```python
check["Contactor"]=0
```

```python
check.info()
```

```python
check
```

```python
for i in check.index:
    if abs(check.loc[i,"Gen_error"]) < 50 and abs(check.loc[i,"Load_error"]) < 50:
        pass
    elif abs(check.loc[i,"Gen_error"]) > 50 and abs(check.loc[i,"Load_error"]) < 50:
        if check.loc[i,"Gen_error"] > 0: #Forecast > Real --> Batt cannot be charged
            check.loc[i,"Contactor"] = 1
        else:
            check.loc[i,"Contactor"] = 0
    elif abs(check.loc[i,"Gen_error"]) < 50 and abs(check.loc[i,"Load_error"]) > 50:
        if check.loc[i,"Load_error"] > 0:
            check.loc[i,"Contactor"] = 0
        else:
            check.loc[i,"Contactor"] = 1
    elif abs(check.loc[i,"Gen_error"]) > 50 and abs(check.loc[i,"Load_error"]) > 50:
        if check.loc[i,"Gen_error"] > 0 and check.loc[i,"Load_error"] > 0:
            pass
        elif check.loc[i,"Gen_error"] > 0 and check.loc[i,"Load_error"] < 0:
            check.loc[i,"Contactor"] = 1
        elif check.loc[i,"Gen_error"] < 0 and check.loc[i,"Load_error"] > 0:
            check.loc[i,"Contactor"] = 0
        elif check.loc[i,"Gen_error"] < 0 and check.loc[i,"Load_error"] > 0:
            check.loc[i,"Contactor"] = 1 
```
